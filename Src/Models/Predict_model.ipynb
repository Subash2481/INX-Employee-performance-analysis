### **1) Load important libraries**

```python
import os
import joblib
import pickle
from dotenv import load_dotenv
import pandas as pd
```

### **2) Load the processed dataset**

```python
df = pd.read_csv(r"C:\Users\subas\Downloads\INX_EMPLOYEE_PERFORMANCE_ANAYSIS\Data\Processed\INX_EMPLOYEE_PERFORMANCE_ANALYSIS_PROCESSED_DATA.csv")
df
```

```python
#Splitting into dependent and independent
X = df.iloc[:, :-1]
y = df.PerformanceRating
```

### **3) Load saved models**

```python
rf_model = joblib.load("random_forest_model.pkl")
gb_model = joblib.load("gradient_boosting_model.pkl")  # if you have saved GB model

print("✅ Models loaded successfully!")
```

```python
print("Loaded RandomForest type:", type(rf_model))
print("Loaded GradientBoosting type:", type(gb_model))
```

### **4) Predict**

```python
rf_predictions = rf_model.predict(X)
gb_predictions = gb_model.predict(X)
```

```python
# Combine results
```

```python
pred_df = pd.DataFrame({
    'Actual': y,
    'RandomForest_Pred': rf_predictions,
    'GradientBoosting_Pred': gb_predictions
})
```

```python
print(pred_df)
```

### **5) Save predictions**

```python
pred_df.to_csv('predictions_output.csv', index=False)
print("\n✅ Predictions completed and saved as 'predictions_output.csv'")
```
